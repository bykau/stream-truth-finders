<h1>Proof of problem(Bayesian algorithm)</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys

%matplotlib inline
```

```python
print 'Python version ' + sys.version
print 'Pandas version ' + pd.__version__
```

```python
st = pd.read_csv('proof_prob_output.csv')
```

```python
prb_dist_norm = [dist/10. for dist in st['dist_metric_mean']]
std_norm = [std/10. for std in st['dist_metric_std']]
plt.plot(st['number_of_swaps'], prb_dist_norm, 'bo-', label='Bayesian algorithm')
plt.errorbar(st['number_of_swaps'], prb_dist_norm, yerr=std_norm)
plt.xlabel('Number of swaps', fontsize=14)
plt.ylabel('Probabilistic distance', fontsize=14)
plt.savefig("proof_exampl.svg")
```
